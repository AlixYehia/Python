````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Use a raw string to prevent Python from interpreting backslashes as escape characters
dataset = pd.read_csv(r'C:\Users\Lenovo\Desktop\4.csv')

# The dataset.head() function is used to display the first few rows of the dataset. 
# When you call dataset.head(), it will print the top 5 rows of the DataFrame by default. 
# This is useful for quickly examining the structure and contents of the dataset to get an initial understanding of the data.
# You can also specify the number of rows you want to display by passing an integer argument to the head() function.
# For example, dataset.head(10) will display the first 10 rows of the dataset.
print(dataset.head())

# The line x = dataset.iloc[:, 0] selects all the rows of the first column (index 0) from the dataset using the iloc method. 
# Here's a breakdown of what each component does:
# dataset: The pandas DataFrame that contains the data loaded from the CSV file.
# iloc: Stands for "integer location-based indexing," which is used to select data by position or index. 
# It allows you to select specific rows and columns by their integer positions.
# [:, 0]: The first colon : indicates that you want to select all the rows from the DataFrame. 
# The 0 after the comma specifies that you want to select the first column (column at index 0).
# [:, :-1]: The first colon : indicates that you want to select all the rows from the DataFrame. 
# The :-1 after the comma specifies that you want to select all the columns except for the last one. 
# This is a way to exclude the target variable (dependent variable) from the feature set.
x = dataset.iloc[:, 0].values
print(x)
y = dataset.iloc[:, 1].values            # negelect last column which is price      .values to remove headers
print(y)


'''

Before reshaping:

```python
import numpy as np

# Example 1D array
x_before = np.array([1, 2, 3, 4, 5])

# Display the shape of x_before
print("Shape of x_before:", x_before.shape)
```

Output:

```
Shape of x_before: (5,)
```

After reshaping:

```python
# Reshape to a 2D array
x_after = x_before.reshape(-1, 1)

# Display the shape of x_after
print("Shape of x_after:", x_after.shape)
```

Output:

```
Shape of x_after: (5, 1)
```

In this example, the original `x_before` array has a shape of `(5,)`, 
indicating that it's a 1D array with 5 elements. After using the `reshape` function, 
the resulting `x_after` array has a shape of `(5, 1)`, indicating that it's a 2D array with 5 rows and 1 column.
'''
x=dataset.iloc[:,:-1].values
# test will be 30%, if randem state is 1 the portion that is divided, 
# in x and y will always be the same after each run and wont change
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

model = LinearRegression()
model.fit(x_train, y_train)     # will give x and y train points to linear regression to draw best line

y_pred = model.predict(x_test)  # output y from trained model if gave x with values. will store them in y_predict

plt.scatter(x_train, y_train, color='red')      # plots the dots that the line will try to follow (real values)
plt.plot(x_train, model.predict(x_train), color='blue')    # I plot the x_trained and the predicted output y from x_trained
plt.xlabel('Price')
plt.ylabel('Volume')
plt.title('Volume vs Price')
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Use a raw string to prevent Python from interpreting backslashes as escape characters
dataset = pd.read_csv(r'C:\Users\Lenovo\Desktop\4.csv')

# Display the first few rows of the dataset
print(dataset.head())

# Extract the feature and target variables
x = dataset.iloc[:, 0].values
print(x)
y = dataset.iloc[:, 1].values
print(y)

# Reshape the arrays to 2D arrays
x = x.reshape(-1, 1)
#y = y.reshape(-1, 1)     dont really need to reshape y since it will be treated as a single value for each value of x

# x_before_shaping = np.array([1, 2, 3, 4, 5])
# output will be (5,) 1D array of 5 elments
# after reshape output will be 2D array (5,1) of 5 elments and 1 column



# Split the data into training and testing sets

# test willl be 30%, if randem state is 1 the portion that is divided, 
# in x and y will always be the same after each run and wont change
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Create and train the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Visualize the data and the linear regression model
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, model.predict(x_train), color='blue')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.title('Volume vs Price')
plt.show()

# Visualize the test set
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_pred, color='blue')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.title('Test Set: Volume vs Price')
plt.show()

# Calculate the error equation (MSE)
error = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {error}")
#In this version, the f before the string allows you to directly insert the value of the variable sme into the string using curly braces {}.

# Equavlient method

# Calculate the Squared Mean Error (SME) using a for loop
n = len(y_test)
sme_sum = 0

for i in range(n):
    sme_sum += (y_test[i] - y_pred[i])**2

sme = sme_sum / n
print(F"Mean Square Error using for loop: {sme}")

# Display the coefficients of the linear regression line
b0 = model.intercept_
b1 = model.coef_[0]
print(f"Coefficient b0: {b0}")
print(f"Coefficient b1: {b1}")

# Sample prediction for a given input value of x
input_value = 507808.444        #that is row 18 column 0 in the given 4.csv file

'''
The double square brackets are used to ensure that the input value is in the form of a 2D array. 
This is required because the predict method in scikit-learn expects a 2D array as input for making predictions.
'''
predict_a_value=float(input("enter a volume to predict its price: "))
predicted_output = model.predict([[predict_a_value]])
print(f"Predicted Price for Volume {predict_a_value}: {predicted_output[0]}\n")


# Calculate the absolute error using a for loop
absolute_errors = []
for i in range(len(y_test)):
    absolute_errors.append(abs(y_test[i] - y_pred[i]))

# Print the absolute errors
print("Absolute Errors:")
for i in range(len(absolute_errors)):
    print(f"Absolute Error {i + 1}: {absolute_errors[i]}")
````
